```python
import pandas as pd
import json

#read from jsonlines in dataset.txt
with open("dataset.txt", "r") as f:
    lines = f.read().split("\n")
    lines = [json.loads(line) for line in lines if line]
    df = pd.DataFrame(lines)


#add data from dataset_saved.txt
with open("dataset_saved.txt", "r") as f:
    lines = f.read().split("\n")
    lines = [json.loads(line) for line in lines if line]
    df_saved = pd.DataFrame(lines)
df = pd.concat([df, df_saved])
```

```python
#read from csv in spreadsheet_dataset.csv
df2 = pd.read_csv("spreadsheet_dataset.csv")
```

```python
df['correct_options']
```

```python
#convert nans to [0], leave arrays untouched
df['correct_options'] = df['correct_options'].apply(lambda x: [0] if type(x)!='list' else x)
df.head()
```

```python
df2.columns
```

```python
#concat interaction histories
df2['interaction']=df2[df2.columns[0]]+ '\n' +df2[df2.columns[1]]
```

```python
df2.head()
```

```python
#put list of cols 1,2,3,4,5 into 'options`
df2['options'] = df2[['1','2','3','4','5']].apply(lambda x: x.tolist(), axis=1)
df2.iloc[0]['options']

#remove nans from options
df2['options'] = df2['options'].apply(lambda x: [i for i in x if not pd.isnull(i)])
df2.iloc[0]['options']
```

```python
df2['correct_options_list'] = df2['Correct answer'].apply(lambda x: x.split(' and '))
df2['correct_options'] = df2['correct_options_list'].apply(lambda x: [int(i) -1 for i in x if not pd.isnull(i)])
df2.head()
```

```python
all = pd.concat([df, df2])[['interaction', 'options', 'correct_options']]
option_dicts = []
for _, entry in all.iterrows():
    option_dict={}
    for i, option in enumerate(entry['options']):
        option_dict.update({option: (i in entry['correct_options']) })
    option_dicts.append(option_dict)
all['option_dicts'] = option_dicts
all.head()

```

```python
all['option_dicts'].iloc[0]
```

```python
#if two rows have same interactions, merge their option_dicts
merged = []
for interaction in all['interaction'].unique():
    matches = all[all['interaction'] == interaction]
    new_row = {}
    new_row['interaction'] = interaction
    new_row['option_dicts'] = {}
    for option_dict in matches['option_dicts']:
        new_row['option_dicts'].update(option_dict)
    merged.append(new_row)

merged = pd.DataFrame(merged)
merged.head()


        

```

```python

merged['option_dicts'][0]
merged['correct_options'] = merged['option_dicts'].apply(lambda x: [i for i in x if x[i]])
merged['incorect_options'] = merged['option_dicts'].apply(lambda x: [i for i in x if not x[i]])
merged.head()
```

```python
merged.shape
```

```python
all_pairs = []
for _, entry in merged.iterrows():
    for correct_option in entry['correct_options']:
        for incorrect_option in entry['incorect_options']:
            all_pairs.append({'interaction': entry['interaction'], 'correct_option': correct_option, 'incorrect_option': incorrect_option})

all_pairs = pd.DataFrame(all_pairs)
all_pairs.head()
all_pairs.shape

```

```python
#save to csv
all_pairs.to_csv('all_pairs.csv')
```

```python
#split merged so there's one option per row
one_per_row = []
for _, entry in merged.iterrows():
    for k,v in entry['option_dicts'].items():
            one_per_row.append({'interaction': entry['interaction'], 'option': k, 'correct?': v})
one_per_row = pd.DataFrame(one_per_row)
one_per_row.to_csv('one_per_row.csv')
one_per_row.head()
```
